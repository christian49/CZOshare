```python
variablesTable = []

def fil_table(original, filtered):
    variable = {}
    variable["original"]=original
    variable["filtered"] = filtered
    add_variable(variable)

def add_variable(variable):
  variablesTable.append(variable)
```

```python
df = pd.read_csv("czo-datasets-metadata-2018-03-01c.csv")
key = df['VARIABLES']
keys = []
for i in range(0,len(key)):
    row = key[i].split('|')
    for j in range(0, len(row)):
        k = row[j].lower()
        k = k.replace("-", "")
        k = k.replace("#", "")
        k = k.replace(" ", "")
        k = k.replace("\t","")
        k = k.replace("(m)","")
        k = k.replace("_", "")
        keys.append(k)
print(keys.__len__())
```

```python
df2 = pd.read_csv("czo-datasets-metadata-2018-03-01c.csv")
key2 = df2['VARIABLES']
keys2 = []
for i in range(0,len(key2)):
    row = key2[i].split('|')
    for j in range(0, len(row)):
        k = row[j].lower()
        keys2.append(k)
print(keys2.__len__())
```

```python
d1 = {}
seen=set()
count = 0
for i in range(len(keys)):
    word = keys[i]
    if word not in seen:
        seen.add(word)
        
        for j in range(len(keys)):
            a_duplicate = keys[j]
            if (word == a_duplicate):
                count += 1   

        d1[word] = count
        count =0
        
d_sorted_special_chars_removed = sorted(d1.items(), key=lambda x: (-x[1], x[0]))
print(len(d_sorted_special_chars_removed))
```

```python
thirdkey = []
for firstkey in d_sorted_special_chars_removed: 
    found = False 
    for secondkey in d_sorted_special_chars_included: 
        if firstkey[0].strip() == secondkey[0].strip():
            found = True
    if not found:
        thirdkey.append(firstkey)
print(len(thirdkey))
```
