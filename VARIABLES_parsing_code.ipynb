```python
df2 = pd.read_csv("czo-datasets-metadata-2018-03-01c.csv")
key2 = df2['VARIABLES']
keys2 = []
for i in range(0,len(key2)):
    row = key2[i].split('|')
    for j in range(0, len(row)):
        k = row[j].lower()
        keys2.append(k)
print(keys2.__len__())
```

```python
def uniquewords(keys):
    seen=set()
    count = 0
    d1 = {}
    for i in range(len(keys)):
        word = keys[i]
        if word not in seen:
            seen.add(word)

            for j in range(len(keys)):
                a_duplicate = keys[j]
                if (word == a_duplicate):
                    count += 1   

            d1[word] = count
            count =0
    return d1, seen
```

```python
d1,seen = uniquewords(keys)
d_sorted_special_chars_removed = sorted(d1.items(), key=lambda x: (-x[1], x[0]))
print(len(d_sorted_special_chars_removed))
```

```python
d2,seen2 = uniquewords(keys2)        
d_sorted_special_chars_included = sorted(d2.items(), key=lambda x: (-x[1], x[0]))
print(len(d_sorted_special_chars_included))
```
