```python
import pandas as pd
import re
```

```python
df = pd.read_csv("/home/christian/Downloads/czo-datasets-metadata-2018-03-01c.csv")
key = df['VARIABLES']
keys = []

for i in range(0,len(key)):
    row = key[i].split('|')
        
    for j in range(0, len(row)):
        k = row[j].lower()
      
        keys.append(k)
```

```python
df1 = pd.read_csv("odmCV.csv")
odmcv = []
for i in df1:
    odmcv.append(i)
    
```

```python
odmcv
```

```python
d1 = {}
seen=set()
count = 0
for i in range(len(keys)):
    word = keys[i]
    if word not in seen:
        seen.add(word)
        
        for j in range(len(keys)):
            a_duplicate = keys[j]
            if (word == a_duplicate):
                count += 1   

        d1[word] = count
        count =0
        
d_sorted1 = sorted(d1.items(), key=lambda x: (-x[1], x[0]))
```

```python
variables = sorted(seen)
sep = ','
odmTerm=[]
for i in range(0,len(odmcv)):
    row = odmcv[i].split(sep,1)[0]
    odmTerm.append(row)
```

```python
len(variables)
```

```python
matchTerms=[]
#Looking for variables that have relations with ODM2 terms
for i in range(0,len(odmTerm)):
    pattern = re.compile(re.escape(odmTerm[i]))
    for j in range(0,len(variables)):
        result = pattern.search(variables[j])
        if result:
            matchTerms.append(odmTerm[i])
            break
            
            
        
```

```python
# removing variables that have relation with ODM2 terms, so we end with a list of variables 
matchVariables=[]
for i in range(0,len(matchTerms)):
    pattern = re.compile(re.escape(matchTerms[i]))
    for j in variables[-1::-1]:
        result = pattern.search(j)
        if result:
            matchVariables.append(variables[variables.index(j)])
            del variables[variables.index(j)]


            
```
