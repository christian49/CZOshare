```python
import pandas as pd
import re
```

```python
df = pd.read_csv("czo-datasets-metadata-2018-03-01c.csv")
key = df['VARIABLES']
# czo_id = df['czo_id']
keys = []


for i in range(0,len(key)):
    row = key[i].split('|')
        
    for j in range(0, len(row)):

        k = row[j].lower()
        
        startingSpacesRemoved = re.sub('\s*$','',k)
        endingSpacesRemoved = re.sub('^\s*','',startingSpacesRemoved)
        cleanString = re.sub('[^a-zA-Z\s]+','',endingSpacesRemoved)  
        
        keys.append(cleanString)
```

```python
d1 = {}
seen=set()
count = 0
for i in range(len(keys)):
    word = keys[i]
    if word not in seen:
        seen.add(word)
        
        for j in range(len(keys)):
            a_duplicate = keys[j]
            if (word == a_duplicate):
                count += 1   

        d1[word] = count
        count =0
        
d_sorted1 = sorted(d1.items(), key=lambda x: (-x[1], x[0]))
```

```python
variables = sorted(seen)
df26 = pd.read_csv("CZO_data_parsing/odmCV/variablename.csv")
array=[df26]
odmCV = []
for i in range(0,len(array)):
    column= array[i]['name']
    for j in range(0,len(column)):
        k = column[j].lower()

        startingSpacesRemoved = re.sub('\s*$','',k)
        endingSpacesRemoved = re.sub('^\s*','',startingSpacesRemoved)
        cleanString = re.sub('[^a-zA-Z\s]+','',endingSpacesRemoved)

        odmCV.append(cleanString)
odmCV = sorted(odmCV)
```
