```python
import re
import pandas as pd
pd.set_option('display.max_colwidth', 500)
pd.set_option('display.max_rows', 400)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("/home/christian/Downloads/czo-datasets-metadata-2018-03-01c.csv")
key = df['VARIABLES']
czo_id = df['czo_id']
keys = []
catalog=[]

for i in range(0,len(key)):
    row = key[i].split('|')
        
    for j in range(0, len(row)):
        index = str(i)+'.'+str(j)
        original = row[j]
        k = original.lower()
        
        startingSpacesRemoved = re.sub('\s*$','',k)
        endingSpacesRemoved = re.sub('^\s*','',startingSpacesRemoved)
        cleanString = re.sub('[^a-zA-Z\s]+','',endingSpacesRemoved)  
        
        keys.append(cleanString)
        catalog.append((index,czo_id[i],cleanString,original))
```

```python
df26 = pd.read_csv("odmCV/variablename.csv")
array=[df26]
odmCV = []
for i in range(0,len(array)):
    column= array[i]['name']
    for j in range(0,len(column)):
        k = column[j].lower()

        startingSpacesRemoved = re.sub('\s*$','',k)
        endingSpacesRemoved = re.sub('^\s*','',startingSpacesRemoved)
        cleanString = re.sub('[^a-zA-Z\s]+','',endingSpacesRemoved)

        odmCV.append(cleanString)
    
```

```python
len(odmCV)
```

```python
d1 = {}
seen=set()
count = 0
for i in range(len(keys)):
    word = keys[i]
    if word not in seen:
        seen.add(word)
        
        for j in range(len(keys)):
            a_duplicate = keys[j]
            if (word == a_duplicate):
                count += 1   

        d1[word] = count
        count =0
        
d_sorted1 = sorted(d1.items(), key=lambda x: (-x[1], x[0]))
```

```python
variables = sorted(seen)
variables
```

```python
matchTerms=[]
#Looking for variables that have relations with ODM2 terms
for i in range(0,len(odmCV)):
    pattern = re.compile(re.escape(odmCV[i]))
    for j in range(0,len(variables)):
        result = pattern.search(variables[j])
        if result:
            matchTerms.append(odmCV[i])
            break
            
            
        
```

```python
# removing variables that have relation with ODM2 terms, so we end with a list of variables 
matchVariables=[]
for i in range(0,len(matchTerms)):
    pattern = re.compile(re.escape(matchTerms[i]))
    for j in variables[-1::-1]:
        result = pattern.search(j)
        if result:
            matchVariables.append(variables[variables.index(j)])
            del variables[variables.index(j)]


            
```
