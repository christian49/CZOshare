```python
import pandas as pd
import re
pd.set_option('display.max_colwidth', 100)
pd.set_option('display.max_rows', 4000)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("table.csv")
key = df['Variable']
unitss = df['Unit']
czo_id = df['czo_id']
disciplines = df['Discipline']
variables = []
df1 = pd.DataFrame(columns=['Variable','Unit','Aggregation','czo_id','Discipline'])

for i in range(0,len(key)):
    k = key[i]
    print(k)
    startingSpacesRemoved = re.sub('\s*$','',k)
    endingSpacesRemoved = re.sub('^\s*','',startingSpacesRemoved)
    pat = re.compile(" ")
    
    variables.append((pat.split(endingSpacesRemoved),czo_id[i],disciplines[i],unitss[i]))
```

```python
var = ""
unt = ""
flag = 0
# units = ['degress']
units=['avg','average','minimum','minimumtimeof','std','stddev','stdev','standarddeviation','standarddeviationof','standarddeviationofmean','max','maximum','maxtimeof','median','medianof','medianofmaximum','medianmaximum','medianofmean','medianofminimum','mediantimeofmaximum','mediantimeofminimum','meanofmaximum','meanofmean','meanofminimum','meanofstandarddeviation','meanstandarddeviationof','vectormean','scalarmean','timeofminimum','timeofmaximum','sum','sumof','total','count']
x = 1
def tablePrint(var=" ",unt="NaN"):
        global x
        global i
        df1.loc[i,'Unit'] = variables[i][3]
        df1.loc[i,'Variable'] = var
        df1.loc[i,'Aggregation'] = unt
        df1.loc[i,'czo_id'] = variables[i][1]
        df1.loc[i,'Discipline'] = variables[i][2]
        x+=1
for i in range(0,len(variables)):
    row = variables[i][0]
    rowLength = len(row)
    
    for j in range(len(units)):
        unit = units[j]
        unitsLength = len(units)
        var = ""
        unt = ""
        for k in range(0,rowLength):
            word = row[k]
            if word == unit:
                unt = word
                flag = 1
                
            elif var == "":
                var = word
            else:
                var = var+" "+word
                
        if flag != 0:
            tablePrint(var,unt)
            flag = 0
            break
        if j == unitsLength-1:
            tablePrint(var)
```
